```python
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import numpy as np

fig, axs = plt.subplot_mosaic([['a','b'],['c','d']], constrained_layout=True)

c=0
lab="abcd"
for suff1,suff in [['c',''],['','_d']]:
    dt=0.1
    for col,ylab in [[4,'n0'], [5,'dc+H.c.']]:
        ax=axs[lab[c]]
        ax.set_xlabel("time")
        ax.set_ylabel(f'|<{ylab}>-exact|')
        ax.set_yscale("log")
        if lab[c]=='a': ax.set_title('discrete interaction picture (bath defined by gs2)')
        if lab[c]=='c': ax.set_title('old Slater phase approach')
        for L in [24,52,100,200]: 
            lt='-'
            n00=np.loadtxt(f"data/dt{dt}/free/free_L{L}.txt", skiprows=1)
            #plt.plot(n0[:,0],[abs(x-0.5) for x in n0[:,4]], lt, label=f"dt={dt} L={L}")
            #plt.plot(n0[:,0],[abs(x-cdRef) for x in n0[:,5]], lt, label=f"dt={dt} L={L}")
        
        
            lt='--'
            n0=np.loadtxt(f"data/dt{dt}/tdvp_noip{suff1}_U0{suff}/L{L}/irlm_no_L{L}.txt", skiprows=1)
            #plt.plot(n0[:,0],[abs(x-0.5) for x in n0[:,4]], lt, label=f"dt={dt} L={L}")
            ti=min(len(n0),len(n00))
            ax.plot(n0[:,0],[abs(x) for x in n0[:ti,col]-n00[:ti,col]], lt, label=f"noip{suff} dt={dt} L={L}")
            
            
        
        ax.set_xlim(0,100)
        ax.set_ylim(1e-5,2e-3)
        ax.legend()
        c=c+1
```
